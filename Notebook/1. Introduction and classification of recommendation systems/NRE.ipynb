```python
links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
new_movies = ratings[ratings['rating'] == 5]
new_movies.head()
```

```python
# movies_with_rating = movies.join(ratings, on='movieId', how='inner', rsuffix='r')
movies_with_rating = movies.join(ratings.set_index('movieId'), on='movieId', how='inner')
# movies_with_rating = movies.merge(ratings, on='movieId')
movies_with_rating.head()
```

```python
# movies_with_rating.groupby('title')[['rating']].mean().sort_values('rating', ascending=False).head(20)
movies_with_rating.groupby('title')[['rating']].sum().sort_values('rating', ascending=False).head(20)
# movies_with_rating.groupby('title')[['rating']].count().sort_values('rating', ascending=False).head(20)
```

```python
ratings_sum = movies_with_rating.groupby('title')[['rating']].sum()
movies_with_tags = movies.join(tags.set_index('movieId'), on='movieId')
```

```python
movies_with_tags.head()
```

```python
tags_count = movies_with_tags.groupby('title')[['tag']].count()
```

```python
tags_count['tag'] = tags_count['tag'] + 1
```

```python
tags_count.head()
```

```python
result_df = ratings_sum.join(tags_count)
```

```python
result_df.head()
```

```python
result_df['res'] = result_df['rating'] * result_df['tag']
```

```python
result_df.sort_values('res', ascending=False).head(25)
```
