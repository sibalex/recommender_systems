### HW Постройте топ фильмов в категориях Action и Comedy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = (14,7)

import warnings
warnings.simplefilter('ignore')

sns.set()
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
# links = pd.read_csv('links.csv') # Для решения текущего задания файл не понадобится.
```

```python
def top_movies(movies=movies, 
               ratings=ratings, 
               tags=tags, 
               top=10, 
               genre='Action', 
               visualization=False):

    """
    Параметры необходимые для запуска функции(top_movies) и краткое описание:

    movies - файл содержит:
        'movieId' (уникальный идентификатор фильма), 
        'title' (Название фильма), 
        'genres' (Жанр фильма).

    ratings - файл содержит:
        'userId' (уникальный идентификатор пользователя),
        'movieId' (уникальный идентификатор фильма),
        'rating' (рейтинг к фильму, от пользователя),
        'timestamp' (время).

    tags - файл содержит:
        'userId' (уникальный идентификатор пользователя),
        'movieId' (уникальный идентификатор фильма),
        'tag' (краткая заметка о фильме).

    top - необходимо указать значение, отображение элементов сортировки (Топ искомой выборки)

    genre - интересующий жанр фильма ['Action', 'Comedy', 'Horror', 'Sci-Fi', ... , n]
    
    visualization - параметр для необходимости визуализации
    """

    movies_ratings = ratings.join(movies.set_index('movieId'), on='movieId') # Объедением фильмы и рейтинг, по id фильма
    
    movies_ratings['genre'] = movies_ratings['genres'].apply(lambda x: genre if genre in x else('Other')) # Зададим условие, если в ['genres'] встречается genre:='жанр' запишем это в отдельнй столбец, в любом другом случае пусть будет 'Other'
    
    movies_ratings = movies_ratings[movies_ratings['genre']==genre] # Убираем лишнее из выборки, кроме интересующего нас жанра
    
    ratings_sum = movies_ratings.groupby(['title', 'genre'])[['rating']].sum() # Группируем Название и Жанр по сумме рейтинга
    
    movies_with_tags = movies.join(tags.set_index('movieId'), on='movieId') # Объединяем movies и tags
    
    tags_count = movies_with_tags.groupby('title')[['tag']].count() # Колличество tags по фильмам
    
    tags_count['tag'] = tags_count['tag'] + 1 # Добавляем +1 на случай 0-вых tags
    
    result_df = ratings_sum.join(tags_count) # Объединяем в результирующую выборку
    
    result_df['res'] = result_df['rating'] * result_df['tag'] # Итоговая формула по топ рейтингу!
    
    result_df.column = ['title', 'genre', 'rating', 'tag', 'res'] # Убираем лишний lvl
    
    result_df = result_df.reset_index() # Сбрасываем индексы
    
    result_df = result_df.sort_values('res', ascending=False).head(top) # Сортируем от большего к меньшему
    
    if visualization == True: # Если нужна визуализация параметр != False
        display(result_df.style.bar())
        print()
        rcParams['figure.figsize'] = (20,7)
        sns.barplot(x=result_df['title'], y=result_df['res'])
        plt.title('Топ {} рейтинг по жанру фильмов: {}'.format(top, genre))
        plt.xlabel('Название фильмов')
        plt.ylabel('Баллы')
        plt.xticks(rotation='30')
        plt.legend(result_df['title'])
        plt.show()
    else:
        return result_df
```

```python
top_movies(top=5, genre='Sci-Fi', visualization=True)
```

```python
top_movies(top=10, genre='Horror', visualization=True)
```

```python
comedy = top_movies(top=5, genre='Comedy', visualization=False)
comedy['rank'] = np.arange(len(comedy)) + 1
comedy = comedy[['rank', 'title']].set_index('rank')
comedy
```
